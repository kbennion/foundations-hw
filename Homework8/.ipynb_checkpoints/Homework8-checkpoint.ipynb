```python
#Which artist has the most songs listed?
print(df['Artist'].value_counts()[:1])
```

```python
#How many hits did the Stones have?
len(df[df['Artist'] == 'the rolling stones'])
```

```python
#How many did The Beatles have?
len(df[df['Artist'] == 'the beatles'])
```

```python
#How many songs are there total? How many years?
(df['Year'].value_counts())
```

```python
#How many songs have the word 'love' in their lyrics?
len(df[df['Lyrics'].str.contains(" love ", na=False)])
```

```python
#How many songs have the word 'no' in their lyrics?
len(df[df['Lyrics'].str.contains(" no ", na=False)])
```

```python
#Did any songs rank for more than one year?
df['Song'].value_counts()[:10]
```
